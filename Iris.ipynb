```python
# A Python program to load the iris data from a given csv file into a dataframe and print the shape of the data, type of the data and first 3 rows.
import pandas as pd
iData = pd.read_csv("iris.csv")
print(iData)
#shape
print("Shape of the data: ",iData.shape)
#type
print("\nData Type: ", type(iData))
# 1st five rows
print("\nFirst 5 rows:\n")
print(iData.head())
```

```python
# A Python program using Scikit-learn to print the keys.
# keys
print("\nKeys of Iris dataset:")
print(iData.keys())
```

```python
# A Python program to get the number of observations, missing values and nan values.
#info
print("\nInfo of Iris data:")
print(iData.info())
```

```python
# A Python program to view basic statistical details like percentile, mean, std etc. of iris data.
# Describing Iris data
print("\nDescription of Iris data:")
print(iData.describe())
```

```python
# A Python program to get observations of each species (setosa, versicolor, virginica) from iris data.
print("\nObservations of each species of Iris data:")
print(iData['Species'].value_counts()) 
```

```python
# A Python program to drop Id column from a given Dataframe and print the modified part. 
print("Original Iris Data:")
print(iData.head())
# Removing 'Id' column
new_iData = iData.drop('Id', axis=1)
print("Iris Data After removing id column:")
print(new_iData.head()) 
```

```python
import matplotlib.pyplot as plt
iData.describe().plot(kind = "area",fontsize = 14, figsize = (14,7), colormap="Accent")
plt.xlabel('Statistics')
plt.ylabel('Value')
plt.title("General Statistics of Iris Dataset")
plt.show()
```
